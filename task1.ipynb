```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt


zip_file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_621128.zip'


with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall('./')


data_file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_621128.csv'
data = pd.read_csv(data_file_path, skiprows=4)  # Skip the first 4 rows which might be metadata


population_2023 = data[['Country Name', '2023']].dropna()


population_2023 = population_2023.sort_values(by='2023', ascending=False)


plt.figure(figsize=(10, 8))
plt.barh(population_2023['Country Name'].head(20), population_2023['2023'].head(20), color='skyblue')
plt.xlabel('Population')
plt.ylabel('Country')
plt.title('Top 20 Countries by Population in 2023')
plt.gca().invert_yaxis()  
plt.show()

```
